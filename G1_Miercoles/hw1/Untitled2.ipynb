```python
# prompt: grafico de barras para 5 variables aleatorias

import matplotlib.pyplot as plt
import numpy as np

# Generar 5 variables aleatorias
variables = np.random.rand(5)

# Etiquetas para las variables
etiquetas = ['Variable 1', 'Variable 2', 'Variable 3', 'Variable 4', 'Variable 5']

# Crear el gráfico de barras
plt.bar(etiquetas, variables)

# Agregar etiquetas y título
plt.xlabel('Variables')
plt.ylabel('Valores')
plt.title('Gráfico de Barras para 5 Variables Aleatorias')

# Mostrar el gráfico
plt.show()
```
