#*Bloque 1 En esta parte del codigo cargamos el Data sets y mostramos cuantas filas y columnas tiene para verificar si estan leyendo el documento correctamente*

```python
import pandas as pd

url = "https://raw.githubusercontent.com/jrudascas/ml-infotep/main/Datasets/houses_medellin.csv"
try:
    df = pd.read_csv(url, encoding="utf-8")
    print("El dataset se cargó correctamente.")
    num_filas, num_columnas = df.shape
    print(f"El dataset tiene {num_filas} filas y {num_columnas} columnas.")
    print("\nColumnas del dataset:")
    print(df.columns.tolist())
except Exception as e:
    print(f"Error al cargar el dataset: {e}")
```

#*Bloque 2 en esta parte limpiamos los datos del data sets antes de buscar duplicados para no generar errores de busueda, eliminamos la primera columna que no sirve para buscar duplicados en la columna porque no tiene ningun valor, colocamos todo en minuscula, eliminamos espacios en blanco y los valores nulos se cambian por cadenas vacias para que la busqueda sea mas efectiva*

```python
df = df.drop(columns=["Unnamed: 0"])
df = df.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
df = df.apply(lambda x: x.str.lower() if x.dtype == "object" else x)
df = df.fillna("")
```

# *Bloque 3 en este bloque creamos una funcion para poder analizar y buscar las filas que tengan todas sus columnas con un valor exacto, es decir duplicadas*

```python
def buscar_duplicados(df):
    duplicados = df[df.duplicated(keep=False)]
    if not duplicados.empty:
        print("\nFilas duplicadas encontradas:")
        conteo_duplicados = duplicados.groupby(df.columns.tolist()).size().reset_index(name='Veces duplicadas')
        for _, row in conteo_duplicados.iterrows():
            print("\nFila duplicada:")
            for columna, valor in row.items():
                if columna != 'Veces duplicadas':
                    print(f"{columna}: {valor}")
            print(f"Veces duplicadas: {row['Veces duplicadas']}")
        num_filas_duplicadas = len(duplicados) - len(conteo_duplicados)
        print(f"\nNúmero total de filas duplicadas: {num_filas_duplicadas}")
    else:
        print("\nNo se encontraron filas duplicadas exactas.")
    return duplicados
```

#*Bloque 4 en este bloque llamamos la funcion de buscar duplicados del bloque 3, en este bloque mostramos los datos duplicados y los conteos de cuantas veces se repiten, ademas eliminamos esas filas duplicadas y nos muestran las filas restantes despues de la eliminacion, luego vulve a buscar filas duplicadas para proceder a eliminarlas*

```python
duplicados = buscar_duplicados(df)

if not duplicados.empty:
    df_sin_duplicados = df.drop_duplicates()
    print(f"\nDespués de eliminar duplicados, el dataset tiene {len(df_sin_duplicados)} filas.")
    print("\nBuscando más filas duplicadas después de la eliminación...")
    duplicados_despues = buscar_duplicados(df_sin_duplicados)

    if not duplicados_despues.empty:
        print("\nAún hay filas duplicadas después de la eliminación.")
    else:
        print("\nNo hay filas duplicadas después de la eliminación.")
else:
    print("\nNo se encontraron filas duplicadas para eliminar.")
```
